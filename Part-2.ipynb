```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('D:/netflix_titles.csv')
data.head()
```

```python
data.date_new=pd.to_datetime(data.date_added)
```

```python
tv=data[data.type=='TV Show']
tv.head()
```

```python
movie=data[data.type=='Movie']
movie.head()
movie.dtypes
```

```python
movie['date_new']=pd.to_datetime(data.date_added)
movie.head()
```

```python
most_productions = movie.date_new.dt.year.value_counts()
most_productions
```

```python
movie.plot(x='release_year',y='',style='o')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.show()
```
